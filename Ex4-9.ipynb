```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
states = np.arange(101)
rewards = np.zeros(101)
policy = np.zeros(101)
rewards[100] = 1
V = np.zeros(101)
ph = .75
```

```python
theta = .0001
while True:
    delta = 0
    for s in states:
        v = V[s]
        actions = list(range(min(s, 100 - s)+1))
        action_values = list(range(min(s, 100 - s)+1))
        
        for i, a in enumerate(actions):
            action_values[i] = ph      * (rewards[s+a] + V[s+a]) + \
                              (1 - ph) * (rewards[s-a] + V[s-a])
        if (0 < s < 100):
            V[s] = max(action_values)
        idx = np.argmax(action_values)
        policy[s] = actions[idx]
        delta = max(delta, abs(v - V[s]))
    fig = plt.plot(V[1:100])
    if delta < theta:
        break
plt.show()
```
